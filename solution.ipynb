### Question 1: 

Given some sample data, write a program to answer the following: [click here to access the required data set](https://docs.google.com/spreadsheets/d/16i38oonuX1y1g7C_UAmiK9GkY7cS-64DfiDMNiR41LM/edit#gid=0)

On Shopify, we have exactly 100 sneaker shops, and each of these shops sells only one model of shoe. We want to do some analysis of the average order value (AOV). When we look at orders data over a 30 day window, we naively calculate an AOV of $3145.13. Given that we know these shops are selling sneakers, a relatively affordable item, something seems wrong with our analysis. 

`a`: Think about what could be going wrong with our calculation. Think about a better way to evaluate this data. 

`b`:  What metric would you report for this dataset?

`c`: What is its value?


```python
import pandas as pd

df1 = pd.read_csv("/Users/victorfrancis/Victor/Professional_Applications/2021_and_After/Shopify-Internship/2019 Winter Data Science Intern Challenge Data Set.csv")

df1.head()
```

```python
df1["AOV"] = df1["order_amount"]/df1["total_items"]
df1.head()
```

```python
df1.AOV.mean()
```

```python
df1.AOV.median()
```

### Solution 1:

`a`: The computation of `AOV` is wrong because `total_items` (number of items ordered) is greater than 1 in some instance, thereby inflating the AOV. The correct AOV can be computed by firstly dividing the `order_amount` by `total_items` before computing the average. However, AOV is itself is a problematic metric.

`b`: **Median Order Value (MOV)**. This is because of the possibility of skewness/outliers. It is well known that mean is a very bad metric for an outlier-prone/skewed dataset. Therefore, `Median Order Value` will be my metric of choice.

`c`: `$153.0`
